```python
class Node:
    def __init__(self, data:int):
        self.data = data
        self.next = None

class Linked_List:
    def __init__(self):
        self.head = None
    
    def append(self, data:int):
        if (not self.head):
            self.head = Node(data)
        else:
            temp = self.head
            while (temp.next):
                temp = temp.next
            temp.next = Node(data)
    
    def traverse(self):
        temp = self.head
        while (temp):
            print(temp.data, "-> ", end="")
            temp = temp.next
        print("NULL")
      
    def search(self, target:int)->int:
        index = -1
        temp = self.head
        while (temp):
            if (temp.data == target):
                return index+1
            temp = temp.next
            index += 1
        return -1

    def insert(self, index:int, data:int):
        if (index >= 0):
            temp = self.head
            self.head = Node(100)
            self.head.next = temp
            temp = self.head
            pre = self.head.next
            curr_index = 0
            while (pre):
                if (curr_index == index):
                    temp.next = Node(data)
                    temp.next.next = pre
                    self.head = self.head.next
                    return
                curr_index += 1
                pre = pre.next
                temp = temp.next
            temp.next = Node(data)
            self.head = self.head.next
        else:
            print("Error index")
        
    def delete(self, index:int):
        if (self.head):
            if (index >= 0):
                temp = self.head
                self.head = Node(100)
                self.head.next = temp
                temp = self.head
                curr_index = -1
                while (temp.next):
                    if (curr_index+1 == index):
                        del_node = temp.next
                        temp.next = temp.next.next
                        del del_node
                        self.head = self.head.next
                        return
                    curr_index += 1
                    temp = temp.next
                print("The index out of the range")
                self.head = self.head.next
            else:
                print("The list is empty")

    def update(self, index:int, data:int):
        if (index >= 0):
            temp = self.head
            curr_index = 0
            while (temp):
                if (curr_index == index):
                    temp.data = data
                    return
                curr_index += 1
                temp = temp.next
            print("The index out of the range")
        else:
            print("Error index")
  
    def count(self, target:int)->int:
        temp = self.head
        count = 0
        while (temp):
            if (temp.data == target):
                count += 1
            temp = temp.next
        return count
    
    def addSort(self, data:int):
        temp = self.head
        self.head = Node(0)
        self.head.next = temp
        temp = self.head
        pre = temp.next
        while (pre and data > pre.data):
            temp = temp.next
            pre = pre.next
        temp.next = Node(data)
        temp.next.next = pre
        self.head = self.head.next


if __name__ == '__main__':
    l1 = Linked_List()
    l1.addSort(1)
    l1.addSort(1)
    l1.addSort(3)
    l1.addSort(2)
    l1.addSort(5)
    l1.addSort(4)
    l1.traverse()
    print(l1.count(1))
```
