```python
import numpy as np
import matplotlib.pyplot as plt 
from sympy.solvers import solve
from sympy import Symbol

def fun1 (x):
    return 215+560*x-100**2
x = Symbol('x')

xr=np.linspace (1, 1200, 80)
ylr=fun1(xr) 
plt.plot(xr, ylr, 'g-')

plt.text(1, 1100, '215+560x-100^2', fontsize=12, color='blue')

plt.xlim (0, 6)
plt.ylim (0, 1200)
plt.show()
```
