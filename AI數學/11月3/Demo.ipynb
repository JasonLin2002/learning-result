```python
import numpy as np
import matplotlib.pyplot as plt 
from sympy.solvers import solve
from sympy import Symbol

def fun1 (x):
    return 7-8*x-2*x**2
x=Symbol('x')

x1=1
x2=2
x3=3

y1=fun1(x1)
y2=fun1(x2)
y3=fun1(x3)

plt.plot(x1, y1, 'go', markersize=10)
plt.plot(x2, y2, 'go', markersize=10)
plt.plot(x3, y3, 'go', markersize=10)

xr = np.linspace(-10, 10, 50)

plt.text(0, 55, 'y=7-8x-2x^2', fontsize=16, color='blue')
y1r = fun1(xr) 

plt.xlim(-10, 10)
plt.ylim(-100, 100)
plt.plot(xr,y1r,'r-')
plt.grid(True)
plt.show()
```
