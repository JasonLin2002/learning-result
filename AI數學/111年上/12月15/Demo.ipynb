```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

data=pd.read_csv('AQI.csv')
data['pm2.5']=data['pm2.5'].fillna(data['pm2.5'].mean())
x_data=np.array(data[['aqi']])
yt=np.array(data[['pm2.5']])
x=np.insert(x_data,0,1.0,axis=1)
plt.scatter(x[:,1],yt,s=10,c='b')
plt.xlabel('AQI',fontsize=14)
plt.ylabel('PM2.5',fontsize=14)
plt.show()
```
