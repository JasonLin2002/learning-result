```python
import numpy as np
import matplotlib.pyplot as plt 
from sympy.solvers import solve
from sympy import Symbol

def fun1(x):
    return -8 + 12 *x + 6 * x ** 2 
x=Symbol('x') 
x1 = 1
x2 = 2
x3 = 3

y1= fun1 (x1)
y2 = fun1 (x2)
y3 = fun1 (x3)

plt.plot(x1, y1, 'go', markersize=10)
plt.plot(x2, y2, 'go', markersize=10)
plt.plot(x3, y3, 'go', markersize=10)

xr = np.linspace (0, 8, 50)

plt.text (2.5, 55, 'y=-8+12x+6x^2', fontsize=12, color='blue')

y1r = fun1 (xr)
plt.plot(xr, y1r, 'b-')
plt.text(1, 8, 'y=12-5x', fontsize=12, color='blue')
plt.xlim(0, 4)
plt.ylim(-20, 120)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.solvers import solve 
from sympy import Symbol

#plt.rcParams['font.sans-serif'] = ['Heiti TC']
def fun1 (x): 
    return -28 +36 * x

x = Symbol('x')
x1 = 1
x2 = 2
x3 = 3

y1 = 10
y2 = 40
y3 = 82

plt.plot (x1, y1, 'go', markersize=8)
plt.plot (x2, y2, 'go', markersize=8)
plt.plot (x3, y3, 'go', markersize=8)

xr = np.linspace (0, 8, 50)

y1r=fun1 (xr)
plt.text(2, 100, 'y -28 +36x', fontsize=16, color='blue')
plt.text(1.3, 10, '(1, 10)', fontsize=14, color='red')
plt.text(2.3, 40, '(2, 40)', fontsize=14, color='red')
plt.text(3.3, 82, '(3, 82)', fontsize=14, color='red')

plt.xlabel('拜訪次數 x (百次)', fontsize=12, color='blue')
plt.ylabel('簽約件數y', fontsize=12, color='blue')

plt.plot(xr, y1r, 'b-')

plt.xlim (0,5)
plt.ylim (0, 150)
plt.grid (True)

plt.show()
```
