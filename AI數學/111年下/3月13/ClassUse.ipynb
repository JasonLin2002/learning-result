```python
import torch
def fun (x1, ml, bl ):
    return ml * x1 + bl
def cost_fun ( yhat1, y1):
    return (yhat1-y1 ) ** 2
xs = torch.tensor( [0, 1, 2, 3, 4, 5, 6, 7])
ys = torch.tensor ([1.86, 1.31, 0.62, 0.33, 0.09, -0.67, -1.23, -1.37])
m = torch.tensor ( 0.9).requires_grad_()
b = torch.tensor ( 0.1 ).requires_grad_()
# Step 1: Forward pass
yhat = fun (xs [7], m, b)
# Step 2: compute yhat with true y to calculate cost C
C = cost_fun ( yhat, ys [7])
# Step 3: Use autodiff to calculate gradient of C w.r.t. parameters
C.backward ()
print( f'partial derivative of C w.r.t. m is {m.grad}')
print( f'partial derivative of C w.r.t. b is {b.grad}')
```
