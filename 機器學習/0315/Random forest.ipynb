```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

heart = pd.read_csv('./heart-disease.csv') # load dataset
X = heart.drop('target', axis=1) # features
y = heart['target'] # target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # split dataset
rfc = RandomForestClassifier(random_state=0) # create model
rfc.fit(X_train, y_train) # train model
y_pred = rfc.predict(X_test) # predict

train_score = rfc.score(X_train, y_train) # show accuracy
test_score = rfc.score(X_test, y_test) # show accuracy
print(f'Train score: {train_score}, Test score: {test_score}') # show accuracy
print('classification report:\n', classification_report(y_test, y_pred)) # show classification report
print('confusion matrix:\n', confusion_matrix(y_test, y_pred)) # show confusion matrix
print(f'accuracy score: {accuracy_score(y_test, y_pred)}') # show accuracy score
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

heart = pd.read_csv('./heart-disease.csv') # load dataset
X = heart.drop('target', axis=1) # features
y = heart['target'] # target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # split dataset

np.random.seed(0) # set seed
best_score = [0, 0] # set best score
for i in range(10, 100, 5): # loop 10 to 100 with step 5
    rfc = RandomForestClassifier(n_estimators=i) # The number of trees in the forest
    rfc.fit(X_train, y_train) # train model
    acc = rfc.score(X_test, y_test) * 100 # show accuracy
    print(f'{i} estimators => accuracy score: {acc:.03f}%') # show accuracy score
    if acc > best_score[1]:
        best_score = [i, acc] # set best score

print(f'best_score[0] decision tree in the model has highest accuracy score: {best_score[1]:.03f}%') 
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pickle

heart = pd.read_csv('./heart-disease.csv') # load dataset
X = heart.drop('target', axis=1) # features
y = heart['target'] # target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # split dataset

np.random.seed(0) # set seed
best_score = [0, 0] # set best score
for i in range(10, 100, 5): # loop 10 to 100 with step 5
    rfc = RandomForestClassifier(n_estimators=i) # The number of trees in the forest
    rfc.fit(X_train, y_train) # train model
    acc = rfc.score(X_test, y_test) * 100 # show accuracy
    print(f'{i} estimators => accuracy score: {acc:.03f}%') # show accuracy score
    if acc > best_score[1]:
        best_score = [i, acc] # set best score
        pickle.dump(rfc, open('./best_rfc.pkl', 'wb')) # save model

print(f'best_score[0] decision trees in the model has highest accuracy score: {best_score[1]:.03f}%') 
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier

heart = pd.read_csv('./heart-disease.csv') # load dataset
X = heart.drop('target', axis=1) # features
y = heart['target'] # target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # split dataset

clf = RandomForestClassifier(random_state=0, n_jobs=-1) # create model
params = {
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 100, 200],
    'n_estimators': [10, 25, 30, 50, 100, 200]
}

grid_search = GridSearchCV(estimator=clf, 
                           param_grid=params, 
                           cv=4, 
                           n_jobs=-1, 
                           verbose=1, 
                           scoring='accuracy') # create grid search

grid_search.fit(X_train, y_train) # train model
print(f'grid search best score: {grid_search.best_score_}') # show out of bag score

clf_best = grid_search.best_estimator_ # get best estimator
print(f'grid search best params: {grid_search.best_params_}') # show best parameters
print(f'grid search best index: {grid_search.best_index_}') # show best index
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier

heart = pd.read_csv('./heart-disease.csv') # load dataset
X = heart.drop('target', axis=1) # features
y = heart['target'] # target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # split dataset

clf = RandomForestClassifier(random_state=0, n_jobs=-1) # create model
params = {
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 100, 200],
    'n_estimators': [10, 25, 30, 50, 100, 200]
}

grid_search = GridSearchCV(estimator=clf, 
                           param_grid=params, 
                           cv=4, 
                           n_jobs=-1, 
                           verbose=1, 
                           scoring='accuracy') # create grid search

grid_search.fit(X_train, y_train) # train model
print(f'grid search best score: {grid_search.best_score_}') # show out of bag score

clf_best = grid_search.best_estimator_ # get best estimator
print(f'best search best params: {clf_best.feature_importances_}') # show best parameters

imp_df = pd.DataFrame({'Varname': X_test.columns, 'Imp': clf_best.feature_importances_}) # create data frame
sorted_feature = imp_df.sort_values('Imp', ascending=False) # sort values
print(f'Sorted features: \n{sorted_feature}') # show sorted features
```
