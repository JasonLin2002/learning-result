```python
import requests
from bs4 import BeautifulSoup
import tkinter as tk
import subprocess

def scrape_website():
    url = url_entry.get()
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    print(str(soup))

def set_font_size():
    font_size = int(font_size_var.get())
    for widget in [lab1, lab2, url_entry, btn1, btn2, btn3, font_size_menu]:
        widget.config(font=("Microsoft JhengHei", font_size))

def open_calculater():
    subprocess.call(['python', '計算機.py'])

window = tk.Tk()
window.title("網頁爬蟲")
window.geometry('800x500')

font_size_var = tk.StringVar()
font_size_var.set("10")

lab1 = tk.Label(window, text="輸入網址:", font=("Microsoft JhengHei", font_size_var.get()))
lab2 = tk.Label(window, text="字體大小:", font=("Microsoft JhengHei", font_size_var.get()))

url_entry = tk.Entry(window, font=("Microsoft JhengHei", font_size_var.get()))

btn1 = tk.Button(window, text="爬取", command=scrape_website, font=("Microsoft JhengHei", font_size_var.get()))
btn2 = tk.Button(window, text="套用", command=set_font_size, font=("Microsoft JhengHei", font_size_var.get()))
btn3 = tk.Button(window, text="計算機", command=open_calculater, font=("Microsoft JhengHei", font_size_var.get()))

font_size_menu = tk.OptionMenu(window,font_size_var, *[i for i in range(24, 40, 2)], font=("Microsoft JhengHei", font_size_var.get()))

lab1.grid(row=0,column=0, padx=10, pady=10)
url_entry.grid(row=1,column=0, padx=10, pady=10)
btn1.grid(row=2,column=0, padx=10, pady=10)

lab2.grid(row=3,column=0, padx=10, pady=10)
font_size_menu.grid(row=3,column=1, padx=10, pady=10)
btn2.grid(row=3,column=2, padx=10, pady=10)

btn3.grid(row=4,column=0, padx=10, pady=10)

for widget in [lab1, lab2, url_entry, btn1, btn2, btn3, font_size_menu]:
    widget.config(font=("Microsoft JhengHei", font_size_var.get()))

window.mainloop()
```

```python
import requests
from bs4 import BeautifulSoup
 
url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
response = requests.get(url=url)
 
soup = BeautifulSoup(response.text, 'lxml')
info_items = soup.find_all('div', 'release_info')
 
for item in info_items:
    name = item.find('div', 'release_movie_name').a.text.strip()
    english_name = item.find('div', 'en').a.text.strip()
    release_time = item.find('div', 'release_movie_time').text.split('：')[-1].strip()
    level = item.find('div', 'leveltext').span.text.strip()
     
    print('{}({}) 上映日：{} 期待度：{}'.format(name, english_name, release_time, level))
```
