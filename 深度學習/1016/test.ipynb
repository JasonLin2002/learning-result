```python
import matplotlib.pyplot as plt
import random

# 從本地端讀取資料
def read_data_from_file(file_path):
    x_values = []
    y_values = []
    with open(file_path, 'r') as file:
        for line in file:
            try:
                x, y = map(int, line.split())
                x_values.append(x)
                y_values.append(y)
            except ValueError:
                # 忽略格式不正確的行
                continue
    return x_values, y_values

# 指定本地端的資料檔案路徑
file_path = 'X_train.txt'

# 讀取資料
x_values, y_values = read_data_from_file(file_path)

# 繪製圖表
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Data from File')

# 添加標題和標籤
plt.title('Data Points Plot from File')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()

# 顯示圖表
plt.show()
```
